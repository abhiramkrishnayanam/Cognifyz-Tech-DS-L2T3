```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter("ignore")
```

```python
df=pd.read_csv("Dataset1.csv")
df.head(2)
```

### 1. Extract additional features from the existing columns, such as the length of the restaurant name or address.

```python
df1=df.copy()
```

```python
# Add new features
df1['Restaurant Name Length']=df1['Restaurant Name'].apply(len) 
#apply(len): Calculates the length of each string in the column. Adds the resulting length as a new column.
df1['Restaurant Address Length']=df1['Address'].apply(len)
```

```python
df1.head(2)
```

### 2. Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables.

```python
df1['Has Table Booking (encoded)']=df1[ "Has Table booking" ].map({'Yes':1,'No':0})
df1['Has Online Delivery (encoded)']=df1[ "Has Online delivery" ].map({'Yes':1,'No':0})
```
